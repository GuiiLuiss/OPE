```python
import pandas as pd
import numpy as np
```

```python
#Importando a base de dados

df = pd.read_csv('C:/Users/guilh/Desktop/OPE - Interative/Dados/Csv/Base_Noverde.csv')
```

```python
#Exluindo a única linha que tem somente valores nulos

df = df.dropna()
```

```python
#Visualizando dataset

df
```

```python
#Visualizando o tipo de dados das colunas

df.dtypes
```

```python
#Tratando campo Gênero de float para int

#Criando nova coluna
df['Genero'] = df['Gênero'].astype('str')

#Substitundo os '.0' para após tranformar em inteiro
df['Genero'] = df['Genero'].apply(lambda x:x.replace('.0',''))

#Transformando em inteiro
df['Genero']=df['Genero'].astype('int64')

df = df.drop('Gênero', axis=1)
```

```python
#Tratando campo Idade de float para int

#Transformando em string
df['Idade']=df['Idade'].astype('str')

#Substitundo os '.0' para após tranformar em inteiro
df['Idade'] = df['Idade'].apply(lambda x:x.replace('.0',''))

#Transformando em inteiro
df['Idade'] = df['Idade'].astype('int64')
```

```python
#Tratando o campo Valor empréstimo

#Criando uma nova coluna e transformando em string
df['ValorEmprestimo']=df['Valor do empréstimo'].astype('str')

#Realizando as substituições para transformar em float
df['ValorEmprestimo'] = df['ValorEmprestimo'].apply(lambda x:x.replace('R$',''))
df['ValorEmprestimo'] = df['ValorEmprestimo'].apply(lambda x:x.replace(',',''))

#Passando para float
df['ValorEmprestimo'] = df['ValorEmprestimo'].astype('float')
```

```python
#Excluindo coluna que já foi tratada e substituida

df = df.drop('Valor do empréstimo', axis=1)
```

```python
#Tratando o campo Valor Final

#Criando uma nova coluna e transformando em string
df['ValorFinal']=df['Valor Final '].astype('str')

#Realizando as substituições para transformar em float
df['ValorFinal'] = df['ValorFinal'].apply(lambda x:x.replace('R$',''))
df['ValorFinal'] = df['ValorFinal'].apply(lambda x:x.replace(',',''))

#Passando para float
df['ValorFinal'] = df['ValorFinal'].astype('float')
```

```python
#Excluindo coluna que já foi tratada e substituida

df = df.drop('Valor Final ', axis=1)
```

```python
#Tratando coluna Quantidade de Parcelas

df['QtdParcelas'] = df['Qtd parcelas'].astype('str')
df['QtdParcelas'] = df['QtdParcelas'].apply(lambda x:x.replace('.0',''))
df['QtdParcelas'] = df['QtdParcelas'].astype('int64')
```

```python
#Excluindo coluna que já foi tratada e substituida

df = df.drop('Qtd parcelas', axis=1)
```

```python
#Tratando o campo Valor Parcelas

#Criando uma nova coluna e transformando em string
df['ValorParcela']=df['Valor Parcelas'].astype('str')

#Realizando as substituições para transformar em float
df['ValorParcela'] = df['ValorParcela'].apply(lambda x:x.replace('R$',''))
df['ValorParcela'] = df['ValorParcela'].apply(lambda x:x.replace(',',''))

#Passando para float
df['ValorParcela'] = df['ValorParcela'].astype('float')
```

```python
 #Excluindo coluna que já foi tratada e substituida

df = df.drop('Valor Parcelas', axis=1)
```

```python
#Tratando o campo Juros

#Transformando em string
df['Juros']=df['Juros'].astype('str')

#Realizando as substituições para transformar em float
df['Juros'] = df['Juros'].apply(lambda x:x.replace('R$',''))
df['Juros'] = df['Juros'].apply(lambda x:x.replace(',',''))

#Passando para float
df['Juros'] = df['Juros'].astype('float')
```

```python
#Transformando em string
df['MotivoEmprestimo']=df['Motivo do empréstimo'].astype('str')
df['Classe'] = df['Classe'].astype('str')
```

```python
#Tratando coluna MotivoEmprestimo

df['MotivoEmprestimo'] = [x.replace('ã','a').replace('é','e').replace('ç','c') for x in df['MotivoEmprestimo']]
```

```python
#Excluindo coluna que já foi tratada e substituida

df = df.drop('Motivo do empréstimo', axis=1)
```
